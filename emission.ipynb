# Carbon Emission

This notebook will contain analytical data about carbon emission around the world and I will predict a couple of years later based on historical data

```python
import pandas as pd
emission_df = pd.read_csv("/kaggle/input/carbon-co2-emissions/Carbon_(CO2)_Emissions_by_Country.csv")
```

```python
emission_df
```

## Kilotons of CO₂ per Country

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for countries in Asia
asia_data = emission_df[emission_df['Region'] == 'Asia']

# Calculate the average Kilotons of CO₂ per country in Asia
asia_avg_co2 = asia_data.groupby('Country')['Kilotons of Co2'].mean().reset_index()

# Sort countries by the average Kilotons of CO₂
asia_avg_co2 = asia_avg_co2.sort_values(by='Kilotons of Co2', ascending=False)

# Create a vertical bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Kilotons of Co2', data=asia_avg_co2, palette='viridis')

# Add title and labels
plt.title('Average Kilotons of CO₂ per Country in Asia')
plt.xlabel('Country')
plt.ylabel('Average Kilotons of CO₂')

# Display the chart
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Metric Tons per Capita

Metric Tons per Capita is a metric to shows average carbon emission on each person in some areas

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter data for countries in Asia
asia_data = emission_df[emission_df['Region'] == 'Asia']

# Calculate the average Kilotons of CO₂ per country in Asia
asia_avg_co2 = asia_data.groupby('Country')['Metric Tons Per Capita'].mean().reset_index()

# Sort countries by the average Kilotons of CO₂
asia_avg_co2 = asia_avg_co2.sort_values(by='Metric Tons Per Capita', ascending=False)

# Create a vertical bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Metric Tons Per Capita', data=asia_avg_co2, palette='viridis')

# Add title and labels
plt.title('Averages Metric Tons Per Capita in Asia')
plt.xlabel('Country')
plt.ylabel('Averages Metric Tons Per Capita')

# Display the chart
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Qatar have the most Metric tons Per Capita in Asia, maybe it is because they have a fossil fuel and industrial that makes emission high, and Qatar has small population so the averages is so high.
